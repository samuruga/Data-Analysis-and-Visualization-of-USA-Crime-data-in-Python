```python
# Data analysing and visualizing the crime data of USA using Python
# Objective question : which state in the USA is the safe or has less crime rate
import numpy as np
import pandas as pd
import csv
from numpy.linalg import inv,det
from matplotlib import pyplot as plt
from matplotlib import style
import seaborn as sns 
# Reading the Data from a csv file 
crime = pd.read_csv('C:/Users/Hi/Desktop/crime.csv')
Data = pd.DataFrame(crime)
print(Data)
```

```python
# Summary of the crime data
describe = crime.describe()
print(describe)
```

```python
# states which has min crime rate with respect to category
a = Data['MURDER'].min()
Murder_min = Data.loc[Data['MURDER']==a,'STATE']
print("The sate where there is min Murder rate is", Murder_min)

b = Data['RAPE'].min()
Rape_min = Data.loc[Data['RAPE']==b,'STATE']
print("The sate where there is min Rape rate is",Rape_min)

d = Data['ASSAULT'].min()
Assault_min = Data.loc[Data['ASSAULT']==d,'STATE']
print("The sate where there is min ASSAULT rate is",Assault_min)
```

```python
#states which has highest theft of personal property, Auto 
f = Data['LARCENY'].max()
Larceny_max = Data.loc[Data['LARCENY']==f,'STATE']
print("The sate where there is max LARCENY rate is",Larceny_max)

g = Data['AUTO'].max()
Auto_max = Data.loc[Data['AUTO']==g,'STATE']
print("The sate where there is max AUTO rate is",Auto_max)
```

```python
# States which has Robbery rate higher than mean value and count of states
med = Data['ROBBERY'].mean()
med_state_robbery = Data.loc[Data['ROBBERY']>med,'STATE']
print("States which has greater than mean Robbery rate\n",med_state_robbery)
print("Count of States greater than mean Robbery rate",med_state_robbery.count())
```

```python
# Bar plot for MURDER
fig=plt.figure(figsize=(10,5))
plt.bar(Data['STATE'],Data['MURDER'], color='G',label='MURDER')
plt.title(' Rate of MURDER in each state')
plt.ylabel('MURDER')
plt.xlabel('STATE')
plt.xticks(rotation=90)
plt.legend(loc='best')
plt.show()
```

```python
### Line plot for Burglary
fig=plt.figure(figsize=(10,5))
#plt.plot(Data['STATE'],Data['BURGLARY'], label='State')
plt.plot(Data['STATE'],Data['BURGLARY'], color='M',label='Burglary', marker ='o')
plt.title(' Rate of Burglary in each state')
plt.ylabel('Burglary')
plt.xlabel('STATE')
plt.xticks(rotation=90)
plt.legend(loc='best')
plt.show()
```

```python
# correlation between Rape and Assault
fig=plt.figure(figsize=(10,5))
plt.scatter(Data['RAPE'],Data['ASSAULT'])
plt.title("Correlation between RAPE and ASSAULT")
plt.ylabel("ASSAULT")
plt.xlabel("RAPE")
plt.xticks(rotation=90)
plt.legend(loc='best')
plt.show()
```

```python
# Pie shart 
fig=plt.figure(figsize=(15,15))
#plt.subplot(1,2,1)
plt.pie(Data['RAPE'], labels= Data['STATE'], autopct ='%1.1f%%')
plt.title('Crime Rate of RAPE')
plt.xticks(rotation=90)
plt.show()
```

```python
# correlation between Larceny, Murder and Rape
fig=plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.regplot(x='RAPE', y="MURDER", data=Data)
plt.title('Correlation between Rape & Murder')
plt.subplot(2,2,2)
sns.regplot(x='LARCENY', y="MURDER", data=Data)
plt.title('Correlation between LARCENY & Murder')
plt.show()
```

```python
# Correlation of crime data
corr=Data.corr()
print(corr)
```

```python
# pass the corr matrix to heatmap object
plt.subplots(figsize=(10,10))
sns.heatmap(corr)
plt.show()
```

```python
# Classifying crime which impacts life and visualizing the top 10 states in both extremes
Data=Data.assign(Life=Data['MURDER']+Data['RAPE']+Data['ASSAULT'])
Data.sort_values(by=['Life'],ascending=[True],inplace=True)
life = Data.head(10)
life = pd.DataFrame(life)
fig=plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(life['STATE'],life['Life'], label='Low Crime', marker = 'o')
plt.title('States with less crime rate impacting Life')
plt.ylabel('Life')
plt.xlabel('STATE')
plt.xticks(rotation=90)
plt.legend(loc='best')
Data.sort_values(by=['Life'],ascending=[False],inplace=True)
life = Data.head(10)
life = pd.DataFrame(life)
plt.subplot(1,2,2)
plt.plot(life['STATE'],life['Life'], label='High Crime', marker = 'o', color = 'r')
plt.title('States with High crime rate impacting Life')
plt.ylabel('Life')
plt.xlabel('STATE')
plt.xticks(rotation=90)
plt.legend(loc='best')
```

```python
# Classifying crime which impacts personal property and visualizing the top 10 states in both extremes
Data=Data.assign(Theft=Data['BURGLARY']+Data['LARCENY']+Data['AUTO']+Data['ROBBERY'])
Data.sort_values(by=['Theft'],ascending=[True],inplace=True)
theft = Data.head(10)
theft = pd.DataFrame(theft)
fig=plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.bar(theft['STATE'],theft['Theft'],label='Low Theft')
plt.title('States with less crime rate impacting personal property')
plt.ylabel('Theft')
plt.xlabel('STATE')
plt.xticks(rotation=90)
plt.legend(loc='best')
plt.subplot(1,2,2)
Data.sort_values(by=['Theft'],ascending=[False],inplace=True)
theft = Data.head(10)
theft = pd.DataFrame(theft)
plt.bar(theft['STATE'],theft['Theft'],label='High Theft', color ='R')
plt.title('States with high crime rate impacting personal property')
plt.ylabel('Theft')
plt.xlabel('STATE')
plt.xticks(rotation=90)
plt.legend(loc='best')
plt.show()
```
